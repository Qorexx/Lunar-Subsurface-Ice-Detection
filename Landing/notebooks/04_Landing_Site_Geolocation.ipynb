```python
crop_size = 1000

top_row = int(best["pixel_row"])
top_col = int(best["pixel_col"])

center_row = top_row + crop_size // 2
center_col = top_col + crop_size // 2

print("Crop Location")
print("-" * 40)

print("Top Left Row :", top_row)
print("Top Left Col :", top_col)

print("Center Row   :", center_row)
print("Center Col   :", center_col)
```

```python
DATA_DIR = Path.home() / "LandingSite" / "Data1"

print(DATA_DIR)

for file in DATA_DIR.iterdir():
    print(file.name)
```

```python
xml_path = DATA_DIR / "ch2_ohr_ncp_20240316T2008014680_d_img_d18.xml"
img_path = DATA_DIR / "ch2_ohr_ncp_20240316T2008014680_d_img_d18.img"

print(xml_path.exists())
print(img_path.exists())
```

```python
text = xml_path.read_text(errors="ignore")

keywords = [
    "LATITUDE",
    "LONGITUDE",
    "UPPER",
    "LOWER",
    "LEFT",
    "RIGHT",
    "CENTER",
    "MINIMUM",
    "MAXIMUM",
    "LINE_PROJECTION_OFFSET",
    "SAMPLE_PROJECTION_OFFSET",
    "MAP_SCALE",
    "MAP_PROJECTION"
]

for line in text.splitlines():
    line_upper = line.upper()
    if any(k in line_upper for k in keywords):
        print(line.strip())
```

```python
# Corner coordinates from XML

UL_LAT = -85.907571
UL_LON = 34.330103

UR_LAT = -85.934206
UR_LON = 33.066466

LL_LAT = -85.095173
LL_LON = 31.275579

LR_LAT = -85.117078
LR_LON = 30.211200

TOTAL_ROWS = 101074
TOTAL_COLS = 12000
```

```python
# Normalize pixel coordinates to [0, 1]

u = center_col / TOTAL_COLS
v = center_row / TOTAL_ROWS

# Bilinear interpolation

landing_lat = (
    (1-u)*(1-v)*UL_LAT +
    u*(1-v)*UR_LAT +
    (1-u)*v*LL_LAT +
    u*v*LR_LAT
)

landing_lon = (
    (1-u)*(1-v)*UL_LON +
    u*(1-v)*UR_LON +
    (1-u)*v*LL_LON +
    u*v*LR_LON
)

print("="*50)
print("FINAL LANDING SITE")
print("="*50)

print(f"Latitude  : {landing_lat:.6f}")
print(f"Longitude : {landing_lon:.6f}")
```

```python
landing_site = {
    "crop_id": int(best["crop_id"]),
    "landing_score": float(best["landing_score"]),
    "latitude": float(landing_lat),
    "longitude": float(landing_lon),
    "shadow_percentage": float(best["shadow_percentage"]),
    "roughness": float(best["roughness"]),
    "max_slope_proxy": float(best["max_slope_proxy"]),
    "pixel_row": int(center_row),
    "pixel_col": int(center_col)
}

import json

with open(PROCESSED_DATA / "landing_site_final.json", "w") as f:
    json.dump(landing_site, f, indent=4)

print("Saved landing_site_final.json")
```
